```python
import pandas as pd
```

```python
df = pd.read_csv("../data/Mobile phone price.csv")
df.head()
```

```python
df.info()
```

```python
categorical_columns = df.select_dtypes(['object', 'category'])
categorical_columns.drop(columns='Price ($)', inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
encoded = encoder.fit_transform(categorical_columns)
```

```python
encoded_df = pd.DataFrame(encoded, columns=categorical_columns.columns)
```

```python
encoded_df['Battery Capacity (mAh)'] = df['Battery Capacity (mAh)']
encoded_df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR

X = encoded_df
y = df['Price ($)'].astype(int, errors='ignore')
y_clean = y.apply(lambda p: str(p).replace("$", "").replace(",","").strip())


scaler = StandardScaler()
scaled = scaler.fit_transform(X)
```

```python
x_train, x_test, y_trian, y_test = train_test_split(X, y_clean, test_size=0.2, random_state=123)
```

```python
model = SVR()
model.fit(x_train ,y_trian)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print(mean_absolute_error(y_pred=y_pred, y_true=y_test))
```

```python
print(r2_score(y_pred=y_pred, y_true=y_test))
```
